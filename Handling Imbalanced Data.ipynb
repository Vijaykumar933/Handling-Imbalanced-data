```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('creditcard.csv')
```

```python
100*sum(df['Class']==1)/sum(df['Class']==0)
```

```python
X = df.drop('Class',axis=1)
```

```python
y = df['Class']
```

```python
## Under Sampling
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
weight = dict({0:1,1:100})
```

```python
rfc = RandomForestClassifier(class_weight=weight)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
rfc.fit(X_train,y_train)
```

```python
y_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
from sklearn.model_selection import KFold,GridSearchCV
```

```python
log_class = LogisticRegression()
grid = {'C':10.0**np.arange(-2,3),'penalty':['l1','l2']}
cv = KFold(n_splits=5,random_state=None,shuffle=False)
```

```python
grid_model = GridSearchCV(estimator=log_class,param_grid=grid,cv=cv,n_jobs=-1,scoring='f1_macro')
```

```python
grid_model.fit(X_train,y_train)
```

```python
y_pred = grid_model.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
y_train.value_counts()
```

```python
from collections import Counter
from imblearn.under_sampling import NearMiss
ns = NearMiss(0.8)
X_train_ns,y_train_ns = ns.fit_resample(X_train,y_train)
print('The no.of classes before fit {}'.format(Counter(y_train)))
print('The no.of classes after fit {}'.format(Counter(y_train_ns)))
```

```python
rfc.fit(X_train_ns,y_train_ns)
```

```python
y_pred = rfc.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
os = RandomOverSampler(0.5)
X_train_ns,y_train_ns = os.fit_resample(X_train,y_train)
print('The no.of classes before fit {}'.format(Counter(y_train)))
print('The no.of classes after fit {}'.format(Counter(y_train_ns)))
```

```python
rfc.fit(X_train_ns,y_train_ns)
```

```python
y_pred = rfc.predict(X_test)
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
